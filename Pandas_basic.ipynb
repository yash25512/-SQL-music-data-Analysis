pandas series

A pandas series is like a column in a table,it is a 1-d array holding data of any type

Series from lists

```python
# string
country=['India','Pakistan','Usa','Nepal','Bhutan']
pd.Series(country)
```

```python
# integers
runs=[13,34,56,78,100]
pd.Series(runs)
```

```python
# custom index
marks=[23,56,78,90,87]
subject=['maths','eng','science','geography','ethics']

pd.Series(marks,index=subject)
```

```python
# setting a name
marks=pd.Series(marks,index=subject,name='yash_marks')
marks
```

Series from dict

```python
marks={
    'maths':78,
    'english':67,
    'sci':56,
    'geo':89
}
marks_series=pd.Series(marks,name='yash_marks')
marks_series
```
